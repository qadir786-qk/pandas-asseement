```python
df = pd.read_csv("C:/Users/Kadir/Downloads/Retail Data.csv")
```

```python
# 1. Total revenue generated across all orders

total_revenue = np.sum(df['Total'])
print(f"1. Total Revenue: {total_revenue}")
```

```python
# 2. Which customer type generates more revenue?

revenue_by_cust = df.groupby('Customer Type')['Total'].sum().idxmax()
print(f"2. Top Customer Type: {revenue_by_cust}")
```

```python
# 3. How does order priority affect revenue?

rev_by_priority = df.groupby('Order Priority')['Total'].sum()
print(f"3. Revenue by Priority:\n{rev_by_priority}")
```

```python
# 4. What is the average profit margin by product category?

avg_margin_cat = df.groupby('Product Category')['Profit Margin'].mean()
print(f"4. Avg Profit Margin by Category:\n{avg_margin_cat}")
```

```python
# 5. What is the most profitable product overall?

most_profitable_prod = df.groupby('Product Name')['Total Profit'].sum().idxmax()
print(f"5. Most Profitable Product: {most_profitable_prod}")
```

```python
# 6. How many days does it usually take to ship an order?
# (Filtering out negative durations caused by data errors)

avg_ship_days = df[df['Shipping Duration'] >= 0]['Shipping Duration'].mean()
print(f"6. Avg Shipping Days: {avg_ship_days:.2f}")
```

```python
# 7. Do longer shipping times impact profit margins? (Correlation)

ship_profit_corr = df[df['Shipping Duration'] >= 0][['Shipping Duration', 'Profit Margin']].corr().iloc[0,1]
print(f"7. Shipping Duration vs Profit Margin Correlation: {ship_profit_corr:.4f}")
```

```python
# 8. Which city brings in the highest revenue?

top_revenue_city = df.groupby('City')['Total'].sum().idxmax()
print(f"8. Highest Revenue City: {top_revenue_city}")
```

```python
# 9. Which account manager generated the most revenue?

top_revenue_manager = df.groupby('Account Manager')['Total'].sum().idxmax()
print(f"9. Top Account Manager: {top_revenue_manager}")
```

```python
# 10. Which shipping mode is most cost-effective?

best_ship_mode = df.groupby('Ship Mode')['Shipping Cost'].mean().idxmin()
print(f"10. Most Cost-Effective Shipping Mode: {best_ship_mode}")
```

```python
# 11. Do higher discounts reduce profits? (Correlation)

discount_profit_corr = df[['Discount %', 'Total Profit']].corr().iloc[0,1]
print(f"11. Discount % vs Total Profit Correlation: {discount_profit_corr:.4f}")
```
